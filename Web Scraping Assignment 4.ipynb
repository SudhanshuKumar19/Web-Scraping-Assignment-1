```python
# Importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
```

# Q5. Scrape the details of trending repositories on Github.com.
# Url = https://github.com/
# You have to find the following details:
# A) Repository title
# B) Repository description
# C) Contributors count
# D) Language used
# Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
```

```python
# Let's first connect with the web driver
driver = webdriver.Chrome(r"C:\Users\Sudhanshu Kumar\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url_5 = "https://github.com/"
driver.get(url_5)
```

```python
#Let's Import required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
```

```python
# Finding the element for trending option using xpath and then clicking it
trending = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
try:
    trending.click()
except ElementNotInteractableException:
    driver.get(trending.get_attribute("href"))
```

```python
# First, Let's create an empty lists to store scraping data

RepositoryTitle = []
RepositoryDescription = [] 
ContributorsCount = [] 
LanguageUsed = []
urls = []
```

```python
# Let's create a function that will find all the elements of different items using xpath
reptitle = driver.find_elements_by_xpath('//article[@class="Box-row"]/h1/a')
descriptions = driver.find_elements_by_xpath('//article[@class="Box-row"]/p')
for i in [reptitle,descriptions]:
    for j in i:
        if i ==reptitle:
            RepositoryTitle.append(j.text)
            urls.append(j.get_attribute('href'))
        if i==descriptions:
            RepositoryDescription.append(j.text)
            
for url in urls:
    driver.get(url)
    driver.implicitly_wait(3)
    divlist=driver.find_elements_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div')
    try:
        ContributorsCount.append(((divlist[-2].text).split())[1])
    except:
        ContributorsCount.append("-")
    try:
        LanguageUsed.append(((divlist[-1].text).split())[1::2])
    except:
        LanguageUsed.append("-")
```

```python
# Import pandas as pd
ques_5 = pd.DataFrame({})
ques_5["Repository Title"] = RepositoryTitle
ques_5["Repository Description"] = RepositoryDescription
ques_5["Contributors Count"] = ContributorsCount
ques_5["Language Used"] = LanguageUsed
ques_5.index+=1

# Details of trending repositories on Github.com
ques_5
```

```python
# Finding the element for Dataset name using xpath
name = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td/table/tbody/tr/td[2]/p/b/a')
for i in name:
    DatasetName.append(i.text)
    
names = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td/table/tbody/tr/td[2]/p/b/a')
for i in names:
    DatasetName.append(i.text)
    
# Finding the element for datatype using xpath 
datatype = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td[2]/p')
for i in datatype:
    DataType.append(i.text)

datatype_1 = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td[2]/p')
for i in datatype_1:
    DataType.append(i.text)
    
# Finding the element for task using xpath
task = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td[3]/p')
for i in task:
    Task.append(i.text)
    
task_1 = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td[3]/p')
for i in task_1:
    Task.append(i.text)
    
# Finding the element for attribute using xpath
att = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td[4]/p')
for i in att:
    AttributeType.append(i.text)
    
attr = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td[4]/p')
for i in attr:
    AttributeType.append(i.text)

# Finding the element for number of instances using xpath
noi = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td[5]/p')
for i in noi:
    NoOfInstances.append(i.text)

noi_1 = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td[5]/p')
for i in noi_1:
    NoOfInstances.append(i.text)
    
# Finding the element for number of attributes using xpath
noa = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td[6]/p')
for i in noa:
    NoOfAttribute.append(i.text)

noa_1 = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td[6]/p')
for i in noa_1:
    NoOfAttribute.append(i.text)
    
# Finding the element for year using xpath
year = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr[2]/td[7]/p')
for i in year:
    Year.append(i.text)
    
yr = driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]/td[7]/p')
for i in yr:
    Year.append(i.text)
```
