# Q1. Write a python program to display all the header tags from 'en.wikipedia.org/wiki/Main_Page'.

```python
from bs4 import BeautifulSoup
import requests
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
soup = BeautifulSoup(page.content)

header = soup.find_all("span", class_="mw-headline")
header_name=[]
for i in header:
    header_name.append(i.text)
header_name

import pandas as pd
Wiki=pd.DataFrame({})
Wiki['HEADERS'] = header_name
Wiki.index+=1
Wiki
    
```

# Q2. Print a python program to display IMDB's Top rated 100 movie's data (i.e Name, IMDB rating, Year of release). 

```python
from bs4 import BeautifulSoup
import requests
page_1 = requests.get("https://www.imdb.com/chart/top")
page_1
```

```python
soup_1 = BeautifulSoup(page_1.content)

name = soup_1.find_all("td", class_="titleColumn")
Name = []
for i in name:
    for j in i.find_all("a"):
        Name.append(j.text.replace("\n",""))
        
rating = soup_1.find_all("td",class_="ratingColumn imdbRating")
Rating = []
for i in rating:
    Rating.append(i.text.replace("\n",""))
        
year = soup_1.find_all("span",class_="secondaryInfo")
Year = []
for i in year:
    Year.append(i.text)

import pandas as pd
IMDB = pd.DataFrame({})
IMDB['NAME'] = Name[0:20]
IMDB['IMDB RATING'] = Rating[0:20]
IMDB['YEAR'] = Year[0:20]
IMDB.index+=1
IMDB
```

```python
IMDB = pd.DataFrame({})
IMDB['NAME'] = Name[0:100]
IMDB['IMDB RATING'] = Rating[0:100]
IMDB['YEAR'] = Year[0:100]
IMDB.index+=1
print(IMDB.to_markdown())
```

# Q5. Write a python program to scrape cricket rankings from 'www.icc-cricket.com'. You have to scrape:

# i) Top 10 ODI teams in men's cricket along with the records for matches, points and rating. 

```python
from bs4 import BeautifulSoup
import requests
page_4 = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page_4
```

```python
soup_4 = BeautifulSoup(page_4.content)

teams = soup_4.find_all("td", class_="table-body__cell rankings-table__team")
Teams = []
for i in teams:
    Teams.append(i.text.replace("\n"," "))

matches = soup_4.find_all("td", class_="table-body__cell u-center-text")
Matches = []
for i in matches:
    Matches.append(i.text)
     
points = soup_4.find_all("td", class_="table-body__cell u-center-text")
Points = []
for i in points:
    Points.append(i.text)
                 
rating = soup_4.find_all("td", class_="table-body__cell u-text-right rating")
Rating = []
for i in rating:
    Rating.append(i.text)
    
import pandas as pd
ODI = pd.DataFrame({})
ODI['TEAMS'] = Teams[0:10]
ODI['MATCHES'] = Matches[0:20:2]
ODI['POINTS'] = Points[1:20:2]               
ODI['RATING'] = Rating[0:10]
ODI.index+=1
ODI
```

```python
# Using Dictionary to get the required data
```

```python
Rows = soup_4.find_all("td")

rank, teams, matches, points, rating = [],[],[],[],[]

dict_ = {0:rank, 1:teams, 2:matches, 3:points, 4:rating}

for i,j in enumerate(Rows):
    dict_[i%5].append(j.text.replace("\n"," ").strip())
    
dict_1 = {
          'RANK' : rank[:10],
          'TEAMS' : teams[:10],
          'MATCHES' : matches[:10],
          'POINTS' : points[:10],
          'RATING' : rating[:10]
                                     }
import pandas as pd
ODI_1 = pd.DataFrame(dict_1)
ODI_1.index+=1
ODI_1
```

# Q7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
from bs4 import BeautifulSoup
import requests
HEADERS = {'User-Agent': 'PostmanRuntime/7.28.4'}
page_10 = requests.get("https://www.amazon.in/s?k=mobiles+under+20000+rupees&ref=nb_sb_noss_1", headers=HEADERS)
page_10
```

```python
soup_10 = BeautifulSoup(page_10.content)

product_name = soup_10.find_all("span", class_="a-size-medium a-color-base a-text-normal")
Product_Name = []
for i in product_name:
    Product_Name.append(i.text)

price = soup_10.find_all("span", class_="a-price-whole")
Price = []
for i in price:
    Price.append(i.text)

image_url = soup_10.find_all("img", class_="s-image")
Image_URL = []
for i in image_url:
    Image_URL.append(i.get('src'))

average_rating = soup_10.find_all("span", class_="a-icon-alt")
Average_Rating = []
for i in average_rating:
    Average_Rating.append(i.text)

import pandas as pd
Amazon_Mobile = pd.DataFrame({})
Amazon_Mobile['PRODUCT NAME (MOBILE)'] = Product_Name[:22]
Amazon_Mobile['PRICE'] = Price[:22]
Amazon_Mobile['IMAGE URL'] = Image_URL[:22]
Amazon_Mobile['AVERAGE RATING'] = Average_Rating[:22]
Amazon_Mobile.index+=1
Amazon_Mobile
```

# Q8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
from bs4 import BeautifulSoup
import requests
page_11 = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YSe_Co4zbMU")
page_11
```

```python
soup_11 = BeautifulSoup(page_11.content)

period = soup_11.find_all("p", class_="period-name")
Period = []
for i in period:
    Period.append(i.text.replace("\n",""))

short = soup_11.find_all("p", class_="short-desc")
Short = []
for i in short:
    Short.append(i.text)

temp = soup_11.find_all("p", class_=["temp temp-high","temp temp-low"])
Temp = []
for i in temp:
    Temp.append(i.text)

desc = soup_11.find_all("div", class_="col-sm-10 forecast-text")
Desc = []
for i in desc:
    Desc.append(i.text)

import pandas as pd
SanFran = pd.DataFrame({})
SanFran['PERIOD'] = Period[:9]
SanFran['SHORT DESCRIPTION'] = Short[:9]
SanFran['TEMPERATURE'] = Temp[:9]
SanFran['FULL DESCRIPTION'] = Desc[:9]
SanFran.index+=1
SanFran
```

# Q9. Write a python program to scrape fresher job listings from 'https://internshala.com/'. It should include job title, company name, CTC and apply date. 

```python
from bs4 import BeautifulSoup
import requests
page_12 = requests.get("https://internshala.com/fresher-jobs")
page_12
```

```python
soup_12 = BeautifulSoup(page_12.content)

job_title = soup_12.find_all("div", class_="heading_4_5 profile")
Job_Title = []
for i in job_title:
    Job_Title.append(i.text.replace("\n",""))

company_name = soup_12.find_all("div", class_="heading_6 company_name")
Company_Name = []
for i in company_name:
    Company_Name.append(i.text.replace("\n","").strip())

ctc = soup_12.find_all("div", class_="item_body")
CTC = []
for i in ctc:  
    CTC.append(i.text)

CTC_New = []
for i in CTC:
    if 'LPA' in i:
        CTC_New.append(i.strip())

applydate = soup_12.find_all("div", class_="item_body")
ApplyDate=[]
for i in range(2,len(applydate),3):
    x = (str(applydate[i]).split('>'))
    ApplyDate.append(x[1].split('<')[0])

import pandas as pd
Fresher_Job = pd.DataFrame({})
Fresher_Job['JOB TITLE'] = Job_Title
Fresher_Job['COMPANY NAME'] = Company_Name
Fresher_Job['CTC'] = CTC_New
Fresher_Job['APPLY DATE'] = ApplyDate
Fresher_Job.index+=1
Fresher_Job
```
